```python
import requests

API_URL = "https://api-inference.huggingface.co/models/monologg/kobert"
headers = {"Authorization": "you keys"}

def query(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.json()
	
output = query({
	"inputs": "Today is a sunny day and I will get some ice cream.",
})

print(output)
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))

def euclidean_distance(A, B):
    return np.sqrt(np.sum((A - B) ** 2))
```

```python
# print(np.array(query({
# 	"inputs": "",
# })))
vec1 = np.array(query({'inputs': ''})).flatten()
vec2 = np.array(query({'inputs': ''})).flatten()

print(f" : {cos_sim(vec1, vec2)}")
print(f"  : {euclidean_distance(vec1, vec2)}")
```
